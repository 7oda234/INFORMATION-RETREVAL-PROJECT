```python
# Sales Distribution by Region
region_sales = df.groupby('Region')['Revenue'].sum()
plt.figure(figsize=(8,8))
region_sales.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Sales Distribution by Region')
plt.ylabel('')
plt.show()
```

```python
# Top 5 Laptop Processors
laptops = df[df['Product'] == 'Laptop']
top_processors = laptops['Processor Specification'].value_counts().head(5)
plt.figure(figsize=(9,6))
top_processors.plot(kind='bar', color='salmon')
plt.title('Top 5 Laptop Processors')
plt.xticks(rotation=45)
plt.ylabel('Count')
plt.grid(True, alpha=0.9)
plt.tight_layout()
plt.show()
```

```python
# Average Price by RAM
ram_price = df.groupby('RAM')['Price'].mean().sort_index()
plt.figure(figsize=(10,6))
ram_price.plot(kind='line', marker='o', color='teal')
plt.title('Average Price by RAM')
plt.xlabel('RAM')
plt.ylabel('Average Price')
plt.grid(True, alpha=0.5)
plt.show()
```
