**import numpy and pandas**

```python
import numpy as np
import pandas as pd
```

**Load the data**

```python
df = pd.read_csv('mobile_sales_data.csv')
```

**Extract columns**

```python
price = df['Price'].values
qty   = df['Quantity Sold'].values
dis   = df['Dispatch Date'].values
ward  = df['Inward Date'].values
ram   = df['RAM'].values
rom   = df['ROM'].values
brand = df['Brand']
R     = df['Region'].values
```

**Average price**

```python
P = np.mean(price)
print(P)
```

**Total units sold**

```python
Q = np.sum(qty)
print(Q)
```

**Most expensive item**

```python
p= np.max(price)
print(p)
```

**Average days in stock**

```python
dw = (dis - ward).mean()
print(dw)
```

**Price volatility**

```python
ps = np.std(price)
print(ps)
```

**Top 10 companies**

```python
B5 = df['Brand'].value_counts().head(10)
print(B5)
```

**Sales by region**

```python
df['Region'].value_counts()
```

**Most common ROM size**

```python
df['ROM'].mode()[0]
```

**Total revenue**

```python
print((price * qty).sum())
```

**Smallest storage size**

```python
print(np.min(rom))
```

**Average units sold per transaction**

```python
print(qty.mean())
```
